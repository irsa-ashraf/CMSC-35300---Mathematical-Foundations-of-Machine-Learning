Clean the tracks and artists dataset so that we have only artists with one genre.

```python
import numpy as np 
import pandas as pd
import os
%matplotlib inline

path = '/Users/irsaashraf/Desktop/UChicago/Fall_22/MathForML/Project/Data/archive'
```

```python
tracks = pd.read_csv(os.path.join(path, 'tracks.csv'))

tracks['id_artists'] = tracks['id_artists'].str.strip('[')
tracks['id_artists'] = tracks['id_artists'].str.strip(']')
tracks['id_artists'] = tracks['id_artists'].str.strip("'")
tracks.head()
```

```python
tracks.rename(columns = {'id': 'id_songs', 'name': 'name_song', 'popularity': 'popularity_songs'}, inplace = True)
```

```python
artists = pd.read_csv(os.path.join(path, 'artists.csv'))
artists.rename(columns = {'id': 'id_artists', 'name': 'name_artists', 'popularity': 'popularity_artists'}, inplace = True)
artists.head()
```

```python
tracks_artists = pd.merge(tracks, artists, left_on = 'id_artists', right_on = 'id_artists')
tracks_artists.head()
```

Checking how many genres we have 

```python
artists['genres'] = artists['genres'].str.strip('[')
artists['genres'] = artists['genres'].str.strip(']')
```

```python
all_genres_lst = []
single_genre_artists_lst = []

for idx, row in artists.iterrows():
    genres_lst = [row['genres']]
    genres_lst = genres_lst[0].split(',')
    if len(genres_lst) >= 1:
        for g in genres_lst:
            if g not in all_genres_lst:
                all_genres_lst.append(g)
    if len(genres_lst) == 1 and genres_lst[0] != '':
        single_genre_artists_lst.append(row['name_artists'])

```

```python
### GETTING ONLY SONGS BY ARTISTS THAT HAVE ONE GENRE ###

single_genre_df = tracks_artists[tracks_artists['name_artists'].isin(single_genre_artists_lst)]
single_genre_df.drop(columns = ['artists'], inplace = True)


# single_genre_df.to_csv(os.path.join(path, 'single_genre_artists.csv'), index = False)
single_genre_df = pd.read_csv(os.path.join(path, 'single_genre_artists.csv'))
single_genre_df.head()
```
